# Data Exploration of Used Cars Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## input of cleaned data
normcars = pd.read_csv("cleaned_cars.csv")
print(normcars)
```

```python
def CorrMatrix(df):
    f = plt.figure(figsize=(19, 15))
    plt.matshow(df.corr(), fignum=f.number)
    plt.xticks(range(df.shape[1]), df.columns, fontsize=14, rotation=45)
    plt.yticks(range(df.shape[1]), df.columns, fontsize=14)
    cb = plt.colorbar()
    cb.ax.tick_params(labelsize=14)
    plt.title('Correlation Matrix', fontsize=16)
    
def CorrPerFeature(df, feature):
    nopr = df.corr()[feature].drop([feature])
    plt.plot(nopr, nopr.index)
    plt.xlabel(str("Correlation with "+feature))
    
def CovMatrix(df):
    f = plt.figure(figsize=(19, 15))
    plt.matshow(df.cov(), fignum=f.number)
    plt.xticks(range(df.shape[1]), df.columns, fontsize=14, rotation=45)
    plt.yticks(range(df.shape[1]), df.columns, fontsize=14)
    cb = plt.colorbar()
    cb.ax.tick_params(labelsize=14)
    plt.title('Covariance Matrix', fontsize=16);
    
def CovPerFeature(df, feature):
    nopr = df.cov()[feature].drop([feature])
    plt.plot(nopr, nopr.index)
    plt.xlabel(str("Covariance with "+feature))
```
